```python
season_runs = matches[['id','season']].merge(
    deliveries,
    left_on='id',
    right_on='match_id'
)

season_runs = season_runs.groupby(
    'season'
)['total_runs'].sum()

season_runs = season_runs.reset_index()

season_runs
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))

sns.lineplot(
    data=season_runs,
    x='season',
    y='total_runs',
    marker='o'
)

plt.title('Total IPL Runs by Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')

plt.show()
```

```python
matches_per_season = matches.groupby('season')['id'].count().reset_index()
matches_per_season
```

```python
season_stats = season_runs.merge(
    matches_per_season,
    on='season'
)
season_stats
```

```python
season_stats = season_stats.rename(
    columns={'id':'matches'}
)
```

```python
season_stats['avg_runs_per_match'] =(
    season_stats['total_runs']
    /
    season_stats['matches']
)    
```

```python
season_stats.sort_values(
    'avg_runs_per_match',
    ascending=False
).head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

sns.lineplot(
    data=season_stats,
    x='season',
    y='avg_runs_per_match',
    marker='o'
)

plt.title('Average Runs Per Match by Season')
plt.show()
```

```python
ipl = matches[['id','season']].merge(
    deliveries,
    left_on='id',
    right_on='match_id'
)
```

```python
team_runs = pd.pivot_table(
    ipl,
    index='batting_team',
    columns='season',
    values='total_runs',
    aggfunc='sum',
)
team_runs
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(
    team_runs,
    annot = False
)
plt.title('Team Runs by Season')
plt.show()
```

Key Insights from IPL Data Analysis
1. Mumbai Indians is the most successful IPL franchise.
MI has won the highest number of matches in IPL history.
This highlights their consistency across multiple seasons.
2. Chennai Super Kings and Kolkata Knight Riders are among the top-performing teams.
CSK ranks second in total wins.
KKR ranks third, showing strong long-term performance.
3. Virat Kohli is the highest run scorer in IPL history.
He has scored more than 8000 runs.
This reflects exceptional consistency and longevity.
4. Yuzvendra Chahal is the leading wicket-taker in IPL history.
He has taken 205 wickets.
His wicket-taking ability has remained consistent across seasons.
5. Caught dismissals dominate IPL cricket.
Most wickets fall through catches.
This indicates the importance of fielding and aggressive batting.
6. Andre Russell has one of the highest strike rates in IPL.
He scores runs at an extremely fast pace.
His impact is more about quick runs than long innings.
7. Chris Gayle is one of the most destructive boundary hitters.
He ranks among the leaders in sixes.
His batting style significantly influenced T20 cricket.
8. Toss has very little impact on match outcome.
Teams winning the toss also won the match only 50.59% of the time.
This suggests team performance matters more than toss advantage.
9. IPL has become more batting-friendly in recent years.
Recent seasons recorded higher total runs.
Modern batting strategies and aggressive intent have increased scoring rates.
10. Some teams show remarkable batting consistency.
Mumbai Indians, Royal Challengers Bangalore, and Kolkata Knight Riders have maintained strong run-scoring performances across multiple seasons.
11. New franchises have quickly become competitive.
Teams like Gujarat Titans have produced strong performances despite entering the league recently.
12. Team participation has evolved over time.
Franchises such as Deccan Chargers, Kochi Tuskers Kerala, Pune Warriors, and Gujarat Lions appeared only during specific IPL periods.
This reflects the changing structure of the tournament.
